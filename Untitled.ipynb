```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
ax = sns.countplot(data=df, x='Laterality', hue='Outcome_Status')

plt.title("Outcome by Eye Affected (Laterality)")
plt.xlabel("Laterality (Left, Right, Both)")
plt.ylabel("Number of Patients")
plt.grid(True, axis='y')

# Move the legend outside the plot
plt.legend(title="Outcome", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Calculate total per Laterality group
total_counts = df['Laterality'].value_counts().to_dict()

# Add percentage labels on top of each bar
for container in ax.containers:
    for bar in container:
        height = bar.get_height()
        if height > 0:
            # Get the Laterality group name from the x position of the bar
            x_pos = bar.get_x() + bar.get_width() / 2
            group = ax.get_xticklabels()[int(x_pos)].get_text()
            total = total_counts[group]
            percent = height / total * 100

            ax.text(
                x_pos,
                height + 3,
                f"{percent:.1f}%",
                ha='center',
                fontsize=9
            )

plt.show()
```

```python
import matplotlib.pyplot as plt

# Define custom colors for the outcomes: blue, orange, green
custom_colors = ['#1f77b4', '#ff7f0e', '#2ca02c']  # blue, orange, green

# Plot the stacked bar chart with custom colors
ax = country_outcomes.plot(kind='bar', stacked=True, figsize=(14, 6), color=custom_colors)

plt.title("Outcome Distribution by Country")
plt.ylabel("Percentage")
plt.xticks(rotation=45)
plt.grid(True, axis='y')

# Move the legend outside the plot
plt.legend(title="Outcome", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Add percentage labels to each bar segment
for idx, country in enumerate(country_outcomes.index):
    cumulative = 0
    for outcome in country_outcomes.columns:
        value = country_outcomes.loc[country, outcome]
        if value > 0:
            plt.text(
                idx,
                cumulative + value / 2,
                f"{value:.1f}%",
                ha='center', va='center',
                fontsize=8
            )
            cumulative += value

plt.show()
```

This above stacked bar chart visualizes the distribution of eye cancer outcomes (In Remission, Deceased, and Active) across different countries. Each bar represents a country, and the colored segments show the percentage of patients in each outcome category.

# Binary Classification

```python
df['Binary_Outcome'] = df['Outcome_Status'].apply(lambda x: 1 if x == 'Deceased' else 0)

Apply one-hot encoding
df_encoded = pd.get_dummies(df.drop(columns=['Outcome_Status']), drop_first=True)

df_encoded['Binary_Outcome'] = df['Binary_Outcome']


X = df_encoded[top_features]
y = df_encoded['Binary_Outcome']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_train_bal, y_train_bal = smote.fit_resample(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

def evaluate_binary_model(model, name):
    model.fit(X_train_bal, y_train_bal)
    y_pred = model.predict(X_test)
    print(f"\n{name} Performance:")
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
    print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
evaluate_binary_model(LogisticRegression(max_iter=1000), "Logistic Regression")
evaluate_binary_model(RandomForestClassifier(), "Random Forest")
evaluate_binary_model(GradientBoostingClassifier(), "Gradient Boosting")
evaluate_binary_model(SVC(), "Support Vector Machine")
evaluate_binary_model(DecisionTreeClassifier(), "Decision Tree")
evaluate_binary_model(KNeighborsClassifier(), "K-Nearest Neighbors")
```

```python
from sklearn.ensemble import VotingClassifier

ensemble = VotingClassifier(
    estimators=[
        ('lr', LogisticRegression(max_iter=1000)),
        ('rf', RandomForestClassifier()),
        ('gb', GradientBoostingClassifier())
    ],
    voting='soft'
)

evaluate_binary_model(ensemble, "Voting Ensemble")
```

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
evaluate_binary_model(xgb, "XGBoost")

```
